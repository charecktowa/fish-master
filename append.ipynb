```python
import pandas as pd

df = pd.read_excel('Tag_Control.xlsx', sheet_name='Sheet1')
df
```

```python
df_1 = pd.read_csv("minka_data/observations-416.csv", low_memory=False)
df_2 = pd.read_csv("minka_data/observations-417.csv", low_memory=False)
df_3 = pd.read_csv("minka_data/observations-418.csv", low_memory=False)

minka = pd.concat([df_1, df_2, df_3], ignore_index=True)
minka.head(3)
```

```python
# Contar cuantas im√°genes hay por especie
species_counts = minka['scientific_name'].value_counts()
species_counts
```

```python
species_counts.index
```

```python
# A√±adir la especie y el total de im√°genes al dataframe de df

for i in range(len(species_counts)):
    if species_counts.index[i].title() in df["Especie"].values:
        print(species_counts.index[i])
        continue

    df.loc[len(df) + 1] = [
        len(df) + 1,
        species_counts.index[i].title(),
        '',
        '',
        species_counts[i],
        '',
    ]  # shifting index
```

```python
# Guardar el dataframe en un archivo Excel
df.to_excel("Tag_Control_2.xlsx", index=False) 
```
