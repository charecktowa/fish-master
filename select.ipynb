```python
import pandas as pd

# Assuming df is already loaded and processed as shown in your code snippet

# Define the fish list
fishes_list = [
    "Aglaophenia",
    "Aplysina aerophoba",
    "Chondrosia reniformis",
    "Cliona celata",
    "Coscinasterias tenuispina",
    "Crambe crambe",
    "Didemnum sp",
    "Dysidea_fragilis",
    "Eudendrium",
    "Ircinia variabilis",
    "Pecten jacobaeus",
    "Reptadeonella violacea",
    "Schizobrachiella sanguinea",
    "Arbacia lixula",
    "Caloria quatrefagesi",
    "Conger conger",
    "Cratena peregrina",
    "Flabellina affinis",
    "Gobius cruentatus",
    "Hexaplex trunculus",
    "Nemesignis banyulensis",
    "Parablennius gattorugine",
    "Parablennius pilicornis",
    "Parablennius tentacularis",
    "Polycera quadrilineata",
    "Serranus cabrilla",
    "Stenopus spinosus",
    "Stramonita haemastoma",
]

# Create lowercase versions of the fish names and their variations
fish_variants = []
for fish in fishes_list:
    # Original name
    fish_variants.append(fish.lower())
    # Replace spaces with underscores
    fish_variants.append(fish.lower().replace(" ", "_"))
    # Replace underscores with spaces
    fish_variants.append(fish.lower().replace("_", " "))


# Function to check if a row contains any fish name in the specified columns
def contains_fish(row, columns):
    for col in columns:
        if pd.isna(row[col]) or row[col] == "":
            continue

        value = str(row[col]).lower()
        for fish in fish_variants:
            if fish in value:
                return True
    return False


# Function to identify which fish is in a row
def identify_fish(row, columns):
    found_fish = []
    for col in columns:
        if pd.isna(row[col]) or row[col] == "":
            continue

        value = str(row[col]).lower()
        for fish in fishes_list:
            fish_lower = fish.lower()
            # Check original, with spaces, and with underscores
            if (
                fish_lower in value
                or fish_lower.replace(" ", "_") in value
                or fish_lower.replace("_", " ") in value
            ):
                if fish not in found_fish:
                    found_fish.append(fish)

    return ", ".join(found_fish) if found_fish else "Unknown"


# Columns to check for fish names
check_columns = ["species_guess", "scientific_name", "common_name"]

# Filter dataframe for rows containing fish names
# First make sure all the columns exist in the dataframe
valid_columns = [col for col in check_columns if col in df.columns]

# Create a filtered dataframe with rows containing fish names
fish_df = df[df.apply(lambda row: contains_fish(row, valid_columns), axis=1)].copy()

# Add a column identifying which fish from the list was found
fish_df["matched_fish"] = fish_df.apply(
    lambda row: identify_fish(row, valid_columns), axis=1
)

# Count occurrences of each fish
fish_counts = {}
for fish in fishes_list:
    fish_lower = fish.lower()
    count = 0

    for _, row in fish_df.iterrows():
        for col in valid_columns:
            if pd.notna(row[col]) and (
                fish_lower in str(row[col]).lower()
                or fish_lower.replace(" ", "_") in str(row[col]).lower()
                or fish_lower.replace("_", " ") in str(row[col]).lower()
            ):
                count += 1
                break  # Count each row only once per fish

    if count > 0:
        fish_counts[fish] = count

# Display summary information
print(f"Found {len(fish_df)} observations containing fish from the list\n")

print("Fish species found and their occurrence counts:")
for fish, count in sorted(fish_counts.items(), key=lambda x: x[1], reverse=True):
    print(f"{fish}: {count} observations")

# Display the entire filtered dataframe
print("\nFiltered dataframe containing all fish observations:")
print(fish_df)

# Optional: Save to a single CSV file if needed
# fish_df.to_csv("all_fish_observations.csv", index=False)
# print("\nSaved all observations to 'all_fish_observations.csv'")
```

```python
# Find all unique fish names in the filtered dataframe
unique_fish = set()

# Show unique fish names
for _, row in fish_df.iterrows():
    for col in valid_columns:
        if pd.notna(row[col]):
            value = str(row[col]).lower()
            for fish in fishes_list:
                fish_lower = fish.lower()
                if (
                    fish_lower in value
                    or fish_lower.replace(" ", "_") in value
                    or fish_lower.replace("_", " ") in value
                ):
                    unique_fish.add(fish)

# Display unique fish names found in the filtered dataframe
print("\nUnique fish names found in the filtered dataframe:")
for fish in sorted(unique_fish):
    print(fish)
```

```python
# Count occurrences of each unique fish name
unique_fish_counts = {fish: 0 for fish in unique_fish}

for _, row in fish_df.iterrows():
    for col in valid_columns:
        if pd.notna(row[col]):
            value = str(row[col]).lower()
            for fish in unique_fish:
                fish_lower = fish.lower()
                if (
                    fish_lower in value
                    or fish_lower.replace(" ", "_") in value
                    or fish_lower.replace("_", " ") in value
                ):
                    unique_fish_counts[fish] += 1

# Display counts of unique fish names
print("\nCounts of unique fish names found in the filtered dataframe:")

# sort
unique_fish_counts = dict(sorted(unique_fish_counts.items(), key=lambda x: x[1], reverse=True))

for fish, count in unique_fish_counts.items():
    print(f"{fish}: {count} observations")
```
