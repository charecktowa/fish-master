```python
import pandas as pd 
import numpy as np
```

```python
# Especies que aparecen en los vídeos

fishes_list = [
    "Aglaophenia",
    "Aplysina aerophoba",
    "Chondrosia reniformis",
    "Cliona celata",
    "Coscinasterias tenuispina",
    "Crambe crambe",
    "Didemnum sp",
    "Dysidea_fragilis",
    "Eudendrium",
    "Ircinia variabilis",
    "Pecten jacobaeus",
    "Reptadeonella violacea",
    "Schizobrachiella sanguinea",
    "Arbacia lixula",
    "Caloria quatrefagesi",
    "Conger conger",
    "Cratena peregrina",
    "Flabellina affinis",
    "Gobius cruentatus",
    "Hexaplex trunculus",
    "Nemesignis banyulensis",
    "Parablennius gattorugine",
    "Parablennius pilicornis",
    "Parablennius tentacularis",
    "Polycera quadrilineata",
    "Serranus cabrilla",
    "Stenopus spinosus",
    "Stramonita haemastoma",
]

fishes_list = [x.lower().replace("_", " ").replace("-", " ") for x in fishes_list]
fishes_list
```

```python
df_1 = pd.read_csv("minka_data/observations-416.csv", low_memory=False)
df_1.head(2)
```

```python
df_2 = pd.read_csv("minka_data/observations-417.csv", low_memory=False)
df_2.head(2)
```

```python
df_3 = pd.read_csv("minka_data/observations-418.csv", low_memory=False)
df_3.head(2)
```

```python
# Unir los dos dataframes
df = pd.concat([df_1, df_2, df_3], ignore_index=True)
df.head(3)
```

```python
# Quitar las columnas que no se van a usar
columnas = [
    "observed_on_string",
    "observed_on",
    "time_observed_at",
    "time_zone",
    "user_id",
    "user_login",
    "user_name",
    "created_at",
    "updated_at",
    "quality_grade",
    "license",
    "url",
    "sound_url",
    "tag_list",
    "description",
    "num_identification_agreements",
    "num_identification_disagreements",
    "captive_cultivated",
    "oauth_application_id",
    "place_guess",
    "latitude",
    "longitude",
    "positional_accuracy",
    "private_place_guess",
    "private_latitude",
    "private_longitude",
    "public_positional_accuracy",
    "geoprivacy",
    "taxon_geoprivacy",
    "coordinates_obscured",
    "positioning_method",
    "positioning_device",
]

df = df.drop(columns=columnas)
df.reset_index(drop=True, inplace=True)
df.head(3)
```

```python
# Count total number of instances in the dataset
total_instances = len(df)
print(f"El dataset contiene {total_instances} observaciones en total.")
```

```python
import re

import matplotlib.pyplot as plt

# Limpieza y estandarización de los nombres científicos
df['cleaned_name'] = df['scientific_name'].fillna('Desconocido')
df['cleaned_name'] = df['cleaned_name'].str.lower().str.strip()
df['cleaned_name'] = df['cleaned_name'].apply(lambda x: re.sub(r'[_\-]', ' ', x))

# Contar ocurrencias de cada especie
species_counts = df['cleaned_name'].value_counts()

# Crear gráfico
plt.figure(figsize=(15, 10))
species_counts.head(20).plot(kind='bar', color='skyblue')
plt.title('Número de Imágenes por Especie (Top 20)')
plt.xlabel('Nombre Científico')
plt.ylabel('Cantidad')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Mostrar el número total de especies
print(f"Número total de especies distintas: {len(species_counts)}")
```

```python
# Buscar especies de fishes_list en el dataframe
found_species = []
not_found_species = []
species_image_counts = {}

for species in fishes_list:
    species_parts = species.split()

    if len(species_parts) == 1:  # Solo género
        # Buscar cualquier especie de este género
        count = df[df["cleaned_name"].str.startswith(species + " ")].shape[0]
        # También buscar coincidencias exactas (por si es una especie sin nombre específico)
        exact_match = df[df["cleaned_name"] == species].shape[0]
        count += exact_match
    else:  # Nombre completo de especie
        # Buscar esta especie específica
        count = df[df["cleaned_name"] == species].shape[0]

    if count > 0:
        found_species.append(species)
        species_image_counts[species] = count
    else:
        not_found_species.append(species)

# Crear mensaje mostrando las especies encontradas
if found_species:
    print(f"Se encontraron {len(found_species)} especies de la lista en el dataset:")
    for species in found_species:
        print(f"- {species}: {species_image_counts[species]} imágenes")
else:
    print("No se encontró ninguna especie de la lista en el dataset.")

if not_found_species:
    print(f"\nNo se encontraron {len(not_found_species)} especies de la lista:")
    for species in not_found_species:
        print(f"- {species}")

# Crear DataFrame para la visualización
species_counts_df = pd.Series(species_image_counts).sort_values(ascending=False)

print(sum(species_image_counts[species]))

# Crear histograma
plt.figure(figsize=(15, 8))
species_counts_df.plot(kind="bar", color="skyblue")
plt.title("Número de Imágenes por Especie (de la lista proporcionada)")
plt.xlabel("Nombre de Especie")
plt.ylabel("Número de Imágenes")
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Buscar especies que contengan "didemnum" en el nombre científico o en species_guess
didemnum_df = df[
    (df['scientific_name'].str.lower().str.contains('didemnum', na=False)) |
    (df['species_guess'].str.lower().str.contains('didemnum', na=False))
]

# Contar cuántas instancias hay de cada especie
if len(didemnum_df) > 0:
    # Usar scientific_name, pero si es NaN usar species_guess
    didemnum_df['species_name'] = didemnum_df['scientific_name'].fillna(didemnum_df['species_guess'])
    species_counts = didemnum_df['species_name'].value_counts()
    
    # Mostrar el número total de instancias
    print(f"Se encontraron {len(didemnum_df)} instancias de especies con 'didemnum' en su nombre:")
    print(species_counts)
    
    # Crear histograma
    plt.figure(figsize=(12, 6))
    species_counts.plot(kind='bar', color='skyblue')
    plt.title('Número de Imágenes por Especies que contienen "didemnum"')
    plt.xlabel('Nombre de Especie')
    plt.ylabel('Número de Imágenes')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
else:
    print("No se encontraron especies con 'didemnum' en su nombre científico o species_guess.")
```

```python
# Crear lista de posibles variantes del nombre
search_terms = ['didemnum lahillei', 'didemnum', 'didemnidae']

# Buscar en varias columnas para cualquier mención de las variantes
results = []
for term in search_terms:
    # Buscar en scientific_name, species_guess y cleaned_name con case insensitive
    matches = df[
        (df['scientific_name'].str.lower().str.contains(term, na=False)) |
        (df['species_guess'].str.lower().str.contains(term, na=False)) |
        (df['cleaned_name'].str.contains(term, na=False))
    ]
    results.append(matches)
    print(f"Encontradas {len(matches)} instancias conteniendo '{term}'")
    
# Combinar resultados y eliminar duplicados
all_results = pd.concat(results).drop_duplicates()

if len(all_results) > 0:
    print(f"\nSe encontraron {len(all_results)} instancias de especies relacionadas con Didemnum:")
    # Mostrar las especies encontradas agrupadas por nombre científico
    species_counts = all_results['scientific_name'].value_counts()
    print(species_counts)
    
    # Mostrar ejemplos de las especies encontradas
    print("\nEjemplos de las especies encontradas:")
    display(all_results[['id', 'scientific_name', 'species_guess', 'cleaned_name']].head(10))
else:
    print("No se encontraron especies relacionadas con Didemnum en el dataset.")
```

```python
# Count unique species in the dataset
unique_species_count = df["cleaned_name"].nunique()
print(f"The dataset contains {unique_species_count} unique species names.")

# Analyze taxonomic levels
# Check how many entries are at species level (contain at least one space)
species_level = df[df["cleaned_name"].str.contains(" ")]["cleaned_name"].nunique()
genus_level = unique_species_count - species_level
print(
    f"Of these, {species_level} are at species level and {genus_level} are at genus level."
)

# Get top 20 species by frequency
top_species = df["cleaned_name"].value_counts().head(20)
print("\nTop 20 species by number of observations:")
print(top_species)

# Visualize species distribution using a pie chart for top 10
plt.figure(figsize=(12, 8))
top_10 = df["cleaned_name"].value_counts().head(10)
others = df["cleaned_name"].value_counts().sum() - top_10.sum()
data = top_10.append(pd.Series({"Others": others}))

plt.pie(data, labels=data.index, autopct="%1.1f%%", startangle=90, shadow=True)
plt.axis("equal")
plt.title("Distribution of Species in the Dataset")
plt.tight_layout()
plt.show()
```
